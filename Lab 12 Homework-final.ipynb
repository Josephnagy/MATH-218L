### Homework Question 1
Consider a line through the origin that makes an angle $\frac{\theta}{2}$ with $x$ axis (that is, the line you get when you rotate the $x$ axis $\frac{\theta}{2}$ anti-clockwise). 
* If you reflect the vector $e_1=\begin{bmatrix}1\\0\end{bmatrix}$ and $e_2=\begin{bmatrix}0\\1\end{bmatrix}$ in this line, what vectors result? Justify your answer with a sketch or otherwise.<br><br>

### Part I ###
If you reflect the vector $e_1=\begin{bmatrix}1\\0\end{bmatrix}$ and $e_2=\begin{bmatrix}0\\1\end{bmatrix}$ in this line, $\begin{bmatrix} \cos\theta \\ \sin\theta\end{bmatrix}$ and $\begin{bmatrix} \sin\theta \\ -\cos\theta\end{bmatrix}$ result.

$\begin{bmatrix}\cos\theta & \sin\theta \\ \sin\theta & -\cos\theta\end{bmatrix}$ $\begin{bmatrix} 1 \\ 0\end{bmatrix}$ = $\begin{bmatrix} \cos\theta \\ \sin\theta\end{bmatrix}$ <br>

$\begin{bmatrix}\cos\theta & \sin\theta \\ \sin\theta & -\cos\theta\end{bmatrix}$ $\begin{bmatrix} 0 \\ 1\end{bmatrix}$ = $\begin{bmatrix} \sin\theta \\ -\cos\theta\end{bmatrix}$

A reflection across the line made by $\frac{\theta}{2}$ above the x-axis is the same as a rotation clockwise by $\frac{\theta}{2}$, reflection across the x-axis, and rotation counterclockwise by $\frac{\theta}{2}$.

$\begin{bmatrix}\cos\frac{\theta}{2} & -\sin\frac{\theta}{2} \\ \sin\frac{\theta}{2} & \cos\frac{\theta}{2}\end{bmatrix}$ $\begin{bmatrix} 1 & 0 \\ 0 & -1\end{bmatrix}$ $\begin{bmatrix}\cos\frac{\theta}{2} & \sin\frac{\theta}{2} \\ -\sin\frac{\theta}{2} & \cos\frac{\theta}{2}\end{bmatrix}$ <br>

$\begin{bmatrix}\cos^2\frac{\theta}{2} -\sin^2\frac{\theta}{2} & 2\cos\frac{\theta}{2}\sin\frac{\theta}{2} \\ 2\cos\frac{\theta}{2}\sin\frac{\theta}{2} & \sin^2\frac{\theta}{2} -\cos^2\frac{\theta}{2}\end{bmatrix}$ <br>

$\begin{bmatrix}\frac{1+\cos\theta}{2}-\frac{1-\cos\theta}{2} & \sin\theta \\ \sin\theta & \frac{1-\cos\theta}{2}-\frac{1+\cos\theta}{2}\end{bmatrix}$ <br>

$\begin{bmatrix}\cos\theta & \sin\theta \\ \sin\theta & -\cos\theta\end{bmatrix}$ <br> <br>

$\begin{bmatrix}\cos\frac{\theta}{2} & \sin\frac{\theta}{2} \\ -\sin\frac{\theta}{2} & \cos\frac{\theta}{2}\end{bmatrix}$ $\begin{bmatrix} 1 & 0 \\ 0 & -1\end{bmatrix}$ $\begin{bmatrix}\cos\frac{\theta}{2} & -\sin\frac{\theta}{2} \\ \sin\frac{\theta}{2} & \cos\frac{\theta}{2}\end{bmatrix}$ <br>
$\begin{bmatrix}\cos^2\frac{\theta}{2} -\sin^2\frac{\theta}{2} & -2\cos\frac{\theta}{2}\sin\frac{\theta}{2} \\ -2\cos\frac{\theta}{2}\sin\frac{\theta}{2} & \sin^2\frac{\theta}{2} -\cos^2\frac{\theta}{2}\end{bmatrix}$ <br>

### Homework Question 3
Let $U_n$ be the first $n$ columns of $U$ (in our case, $U_1$ is the first column, and $U_2=U$, since there are only two columns). Consider the product $U_1^TX$, where $X$ is our normalized (avg = 0) data. Compute it and compare it to $S_1V^T_1$. Explain your result. How would you compute the projection onto the first principal direction using $U_1^TX$ instead of computing $U_1S_1V^T_1$?<br><br>

**Note:** The method this question is far more efficient than the one in the lab, as it requires multiplication of smaller matrices. We will use this idea in the next lab. We can view the product $U_1^TX$ as the position along the first principal axis of a given face. In other words, it's a set of *coordinates*, one for each face.

```python
import numpy as np
faces=np.genfromtxt('measurementsf2019.csv',delimiter=',').T
averagex = np.sum(faces[0])/len(faces[0])
averagey = np.sum(faces[1])/len(faces[1])

normalized = np.zeros_like(faces)
normalized[0] = faces[0] - averagex
normalized[1] = faces[1] - averagey

U,S,Vt = np.linalg.svd(normalized)
print(np.isclose((U[:,:1].T@normalized),np.diag(S)[:1,:1]@Vt[:1]))

print(U.shape)
print(np.diag(S).shape)
print(Vt.shape)
```

$X=USV^T$ <br>
$U^TX = SV^T$ <br>
So, to compute the projection onto the first principal direction, we would compute the product of U[:,:1].T and X

### Homwork Question 4
Hand in your six plots with brief explanations of each.

```python
%matplotlib notebook
import matplotlib.pyplot as plt

def middleaxes(plot):
    # Given a matplotlib plot, this sets the axes to be centered at the origin
    # Returns the plot
    plot.spines['left'].set_position('zero')
    plot.spines['right'].set_color('none')
    plot.yaxis.tick_left()
    plot.spines['bottom'].set_position('zero')
    plot.spines['top'].set_color('none')
    plot.xaxis.tick_bottom()

    return plot
faces=np.genfromtxt('measurementsf2019.csv',delimiter=',')
fig, ax = plt.subplots(3, 2)        # create a figure and m x n subplots in a grid
fig.set_size_inches(10,10)   # set the overall height and width of the figure
ax[0,0].axis([0,90,0,40]) # set min and max for x and y axes
ax[0,0].set_aspect('equal')         # set an 1-1 aspect ratio

#Part 1 - first graph
middleaxes(ax[0,0])
ax[0,0].scatter(faces[:,0],faces[:,1])

#Part 2 - second graph
middleaxes(ax[0,1])
ax[0,1].axis([-15,15,-15,15])

rows,cols=faces.shape
translated = np.zeros((rows,cols))

translated[:,0] = faces[:,0] - averagex
translated[:,1] = faces[:,1] - averagey
ax[0,1].scatter(translated[:,0],translated[:,1])

#Part 3 - third graph
translated = translated.T
U, S, Vt = np.linalg.svd(translated)
rotated = U.T@translated

ax[1,0].axis([-15,15,-15,15])
ax[1,0].set_aspect('equal')
ax[1,0].scatter(rotated[0],rotated[1])
middleaxes(ax[1,0])

#adding arrows
from matplotlib.patches import Arrow
arr = Arrow(averagex,averagey, 7*U[0,0],7*U[1,0], width=0.20,color="red")
arr2 = Arrow(averagex,averagey,3*U[0,1],3*U[1,1], width=0.20,color="red")
ax[0,0].add_patch(arr)  
ax[0,0].add_patch(arr2)

arr3 = Arrow(0,0, 7*U[0,0],7*U[1,0], width=0.20,color="red")
arr4 = Arrow(0,0,3*U[0,1],3*U[1,1], width=0.20,color="red")
ax[0,1].add_patch(arr3)  
ax[0,1].add_patch(arr4)

UTU = U.T@U
arr5 = Arrow(0,0, 7*UTU[0,0],7*UTU[0,1], width=0.20,color="red")
arr6 = Arrow(0,0,3*UTU[1,0],3*UTU[1,1], width=0.20,color="red")
ax[1,0].add_patch(arr5)  
ax[1,0].add_patch(arr6)

#adding ellipses
from matplotlib.patches import Ellipse
width = 2*(1/np.sqrt(rows))*S[0]
height = 2*(1/np.sqrt(rows))*S[1]
E = Ellipse(xy=(0,0),width=2*width,height=2*height,color="orange",alpha=0.2)
ax[1,0].add_patch(E)

angle = np.degrees(np.arccos(U[0,0]))
unrotatedE = Ellipse(xy=(0,0),width=2*width, height=2*height,color="orange",alpha=0.2,angle=-angle)
ax[0,1].add_patch(unrotatedE)

untranslatedE = Ellipse(xy=(averagex,averagey),width=2*width, height=2*height,color="orange",alpha=0.2,angle=-angle)
ax[0,0].add_patch(untranslatedE)

#Part 4 - compressed circle
stdX = S[0]/np.sqrt(rows)
stdY = S[1]/np.sqrt(rows)
middleaxes(ax[1,1])
ax[1,1].scatter(rotated[0]/stdX,rotated[1]/stdY)
ax[1,1].axis([-15,15,-10,10])

width2 = (2*S[0])/(np.sqrt(rows)*stdX)
height2 = (2*S[1])/(np.sqrt(rows)*stdY)
compressedE = Ellipse(xy=(0,0),width=2*width2,height=2*height2,color="orange",alpha=0.2)
ax[1,1].add_patch(compressedE)

#Part 5
def SVDRebuild(U,S,V,k):
    return U[:,:k]@S[:k,:k]@V.T[:k,:]

rebuilt = SVDRebuild(U,np.diag(S),Vt.T,1)
rebuilt[0] += averagex
rebuilt[1] += averagey
middleaxes(ax[2,0])
ax[2,0].scatter(rebuilt[0],rebuilt[1])
ax[2,0].axis([10,80,0,25])
ax[2,0].scatter(faces[:,0],faces[:,1],color="orange",alpha=0.2)

#Part 6
indices = np.random.randint(101,size=5)
ax[2,1].scatter(faces[:,0][indices],faces[:,1][indices],color="red")
ax[2,1].scatter(rebuilt[0][indices],rebuilt[1][indices],color="blue")
for i in range(5):
    arr5 = Arrow(faces[:,0][indices[i]],faces[:,1][indices[i]],rebuilt[0][indices[i]] - faces[:,0][indices[i]],
                 rebuilt[1][indices[i]]-faces[:,1][indices[i]],color="orange",alpha=0.2,width=0.2)
    ax[2,1].add_patch(arr5)
```

ax[0,0]: displays the unmanipulated data. The transparent ellipse is centered at the data's mean and has width and heights that are double their respective standard deviations. The arrows indicate the principle directions. <br>
ax[0,1]: displays the displaced data, which has been translated by the mean in both the x and y directions, so that the center is at the origin. The ellipse remains unchanged. <br>
ax[1,0]: displays the displaced and rotated data, where the principle directions are now the x and y axes.<br>
ax[1,1]: displays the data scaled down to form a circle. <br>
ax[2,0]: the blue data points represent the original data projected onto the first principal direction, and the orange data points represent the original data points. <br>
ax[2,1]: the red data points represent a random sample of the original data, the blue data points are their projections onto the first principal direction. The orange arrow shows their transformation.
